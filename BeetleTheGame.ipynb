```python
from random import randint

def beetle221():
    """Version of beetle game with statistics"""
    L = [0,0,0,0,0,0] # empty list which store our progress
    M = [1,1,2,2,1,6] # when L list reach these values then game stop
    C = [0,0,0,0,0,0] # list to store counters
    rolls = 0 # counts number of rolls
    # Few colours to make our progress in console more transparent
    a = "\033[1;32m" # green
    b = "\033[1;31m" # red
    c = "\033[1;33m" # yellow
    d = "\033[0;37m" # gray/grey
    e = "\033[1;34m" # blue
    z = "\033[0m" # neutral
    print(e+"Welcome to the game!") # inital comment, for start recognition
    print("Let's start. Good luck!"+z)
    keep_going = True # condition to keep game run
    while (keep_going): 
        if L != M:  # comparing lists, if satisfied then roll
            rolls += 1 # adding rolls
            roll = randint(1, 6) # range of rolls that we use to play
            print("You rolled: %i"  % roll)
            if roll == 1: # need to roll 1 to get 2 or 6
                C[0] += 1 # adding particular rolled value to list
                if L[0] < M[0]: # add body if roll 1 and don"t have body yet 
                    L[0] += 1
                    print(a+"You get a body! "
                          "Now you can get a head or legs!"+z)
                elif L[0] == M[0]: # if already have a body then do nothing.
                    print(c+"You already have a body! Roll another value."+z)
            elif roll == 2: # need to roll 2 to get 3, 4 or 5
                C[1] += 1 # adding particular rolled value to list
                if L[0] != M[0]: # if don"t have body yet do nothing
                    print(b+"You can't get a head before you will get a body."
                          " Roll another value."+z)
                elif (L[0] == 1) and (L[1] < M[1]):
                # if have a body then add a head till get 1
                    L[1] += 1 # adding particular rolled value to list
                    print(a+"You get a head! Now you can get an antenna," 
                          " an eye and a mouth!" +z)
                elif L[1] == M[1]: # if already have a head then do nothing.
                    print(c+"You already have a head. Roll another value."+z)
            elif roll == 3: # if 1 & 2 satisfied then can add 3
                C[2] += 1 # adding particular rolled value to list
                if L[0] == 0 or L[1] == 0: # if 1 and 2 not satisfied 
                # then do nothing
                    print(b+"You can't get an antenna before you will get" 
                          " a head. Roll another value."+z)
                elif (L[1] == M[1]) and (L[2] < M[2]):
                # if have a head then add an antennas till get 2 of them
                    L[2] += 1
                    print(a+"You get an antenna!"+z)
                elif L[2] == M[2]: # if already have both antennas
                # then do nothing.
                    print(c+"You already have both antennas."
                          " Roll another value."+z)
            elif roll == 4:
                C[3] += 1 # adding particular rolled value to list
                if L[0] == 0 or L[1] == 0:
                # if 1 and 2 not satisfied then do nothing
                    print(b+"You can't get an eye before you will get head."
                          " Roll another value."+z)
                elif (L[1] == M[1]) and (L[3] < M[3]):
                # if have a body then add a head till get 1
                    L[3] += 1
                    print(a+"You get an eye!"+z)
                elif L[3] == M[3]: # if already have both eyes then do nothing.
                    print(c+"You already have both eyes."
                          " Roll another value."+z)
            elif roll == 5:
                C[4] += 1 # adding particular rolled value to list
                if L[0] == 0 or L[1] == 0:
                # if 1 and 2 not satisfied then do nothing
                    print(b+"You can't get a mouth before you will get head."
                          " Roll another value."+z)
                elif (L[1] == M[1]) and (L[4] < M[4]): # if have head 
                # then add mouth 
                    L[4] += 1
                    print(a+"You get a mouth!"+z)
                elif L[4] == M[4]: # if already have a mouth then do nothing.
                    print(c+"You already have a mouth. Roll another value."+z)
            elif roll == 6:
                C[5] += 1 # adding particular rolled value to list
                if L[0] == 0: # if 1 not satisfied then do nothing
                    print(b+"You can't get a leg before you will get body!"
                          " Roll another value."+z)
                elif (L[0] == M[0]) and (L[5] < M[5]):
                    # if 1 satisfied then can add 6
                    L[5] += 1
                    print(a+"You get a leg!"+z)
                elif L[5] == M[5]: # if already all eyes then do nothing.
                    print(c+"You already have all legs. Roll another value."+z)
            print(d + "Your stack:"+ z, L) # new value pop up if valid
        elif L == M: # 
            print(e+" It was a last roll, well done üêû", "\n",
                  "Player has collected all body parts", L, "\n",
                  "Here is summary of your rolls:", C, "\n", 
                  "In total you rolled your dice", rolls, "times.","\n",
                  "Odd rolls:", C[0] + C[2] + C[4],
                  "Even rolls:", (C[1] + C[3] + C[5]))
            keep_going = False # condition to stop the game          

beetle221()
```
